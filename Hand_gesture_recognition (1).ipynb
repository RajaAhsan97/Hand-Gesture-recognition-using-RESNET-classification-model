```python
from zipfile import ZipFile
ds_file = '/content/splitted sign language data.zip'

with ZipFile(ds_file, 'r') as zip:
  zip.extractall()
print('file extracted...')
```

```python
IMG_WIDTH, IMG_HEIGHT = 112, 112
```

```python
def process_image(input_image):
  #image = tf.image.decode_jpeg(image, channels=3)
  image = tf.image.resize(input_image, [IMG_WIDTH, IMG_HEIGHT])
  #image = tf.cast(image, tf.float32)
  #image = image/255.0
  return tf.keras.applications.resnet50.preprocess_input(image)
  #return image
```

```python
train_dir = 'splitted sign language data _ v2/train'
test_dir = 'splitted sign language data _ v2/test'
valid_dir = 'splitted sign language data _ v2/valid'

train_datagen = ImageDataGenerator(
    #rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    preprocessing_function = process_image)
#    preprocessing_function = tf.keras.applications.resnet50.preprocess_input)
#    preprocessing_function=edge_detection)

test_datagen = ImageDataGenerator(
    preprocessing_function = process_image
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=8,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=8,
    class_mode='categorical'
)

val_generator = test_datagen.flow_from_directory(
    valid_dir,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=8,
    class_mode='categorical'
)
```

```python
input_layer = layers.Input(shape=(IMG_WIDTH, IMG_HEIGHT,3))
model_resnet = resnet50.ResNet50(weights='imagenet', input_tensor=input_layer, include_top=False)
#model_resnet.summary()
```

```python
last_layer = model_resnet.output
flatten = layers.Flatten()(last_layer)
# for embeddings 128 vector
#out = layers.Dense(128, activation='relu')(flatten)
dropout = layers.Dropout(0.5)(flatten)
#output_layer = layers.Dense(len(class_names), activation='softmax')(dropout)
output_layer = layers.Dense(train_generator.num_classes, activation='softmax')(dropout)
```

```python
model = models.Model(inputs=input_layer, outputs=output_layer)
```

```python
for layer in model.layers[:-3]:
  layer.trainable=False
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.00001)
```

```python
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath="checkpoints_input_preprocess/model_epoch_{epoch:02d}_val_accuracy_{val_accuracy:.2f}.keras",  # Path to save the model
    monitor="val_accuracy",  # Metric to monitor (e.g., validation loss)
    save_best_only=True,  # Save only the best model
    save_weights_only=False,  # Save the entire model (architecture + weights)
    mode="max",  # Save the model when the monitored metric is minimized
    verbose=1,  # Print a message when saving the model
)
```

```python
history = model.fit(train_generator,
                    epochs=60,
                    batch_size=8,
                    verbose=True,
                    validation_data=val_generator,
                    callbacks=[checkpoint_callback])
                    #callbacks=[earlystopping])
```

```python
y_true = []
y_pred = []

for i in range(num_batches):
    x, y = test_generator[i]
    y_true.extend(np.argmax(y, axis=1))
    y_pred.extend(np.argmax(model.predict(x), axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)
```

```python
cm = confusion_matrix(y_true, y_pred)
```

```python
class_names = list(test_generator.class_indices.keys())
```

```python
plt.figure(figsize=(20,20))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion matrix")
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)

thresh = cm.max() / 2
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j,i, format(cm[i,j], 'd'),
             horizontalalignment='center',
             color='white' if cm[i, j] > thresh else 'black')

plt.ylabel('True label')
plt.xlabel('Predicted label')

#plt.savefig('DS1_out/real image/RESNET_model_DS1_CED_v6_CM_val.png')
plt.show()
```

```python
from sklearn.metrics import average_precision_score
```

```python
y_pred = model.predict(test_generator)
print(y_pred[:,0].shape)
y_true = []
num_batches = test_generator.num_batches
for i in range(num_batches):
    x, y = test_generator[i]
    y_true.extend(y)
y_true = np.array(y_true)
```

```python
#Initialize precision, recall, and average precision lists
precision = dict()
recall = dict()
average_precision = dict()

num_classes = test_generator.num_classes

# Calculate precision-recall curve for each class
for i in range(num_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_true[:,i], y_pred[:,i])
    average_precision[i] = average_precision_score(y_true[:,i], y_pred[:,i])
```

```python
plt.figure(figsize=(8, 8))
for i in range(num_classes):
    plt.plot(recall[i], precision[i], label='class {0} (area = {1:0.2f})'.format(i, average_precision[i]))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curve for Multi-class Classification')
plt.legend(loc="lower right")
plt.show()

# Calculate and print average precision (micro-averaged)
average_precision["micro"] = average_precision_score(y_true, y_pred, average="micro")
print(f"Average precision (micro-averaged): {average_precision['micro']:.3f}")
```
